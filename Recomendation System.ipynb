```python
# Let us look at the cosine similarity between two movies
```

```python
from scipy import spatial
```

```python
from numpy import dot
from numpy.linalg import norm
```

```python
li1 = list(user_movie_ratings['101 Dalmatians (1996)'])
li2 = list(user_movie_ratings['2001: A Space Odyssey (1968)'])
```

```python
cos_simi = dot(li1,li2)/(norm(li1)*norm(li2))
```

```python
cos_simi
```

```python
def recomendation(movie):
    simi=[]
    mov=[]
    li1 = list(user_movie_ratings[movie])
    for i in movie_d.index:
        li2 = list(user_movie_ratings[i])
        cos = dot(li1,li2)/(norm(li1)*norm(li2))
        simi.append(cos)
        mov.append(i)
    recomd_data = pd.DataFrame({'Movie':pd.Series(mov),'Similarity':pd.Series(simi)})
    return recomd_data.sort_values(by='Similarity',ascending=False).iloc[1:11,:]
```

```python
# Recomendation based on Feature(gerne)  [Content based filtering]
```

```python
movie_dc = movie_d
```

```python
movie_dc.set_index('movie title',inplace=True)
```

```python
movie_dc = movie_d.T
```

```python
movie_dc.drop(index=['movie id','release date','video release date','IMDb URL'], inplace=True)
```

```python
movie_dc = movie_dc.T.drop_duplicates().T
```

```python
movie_dc
```

```python
def content_recom(movie):
    simi=[]
    mov=[]
    li1 = list(movie_dc[movie])
    for i in movie_dc.columns:
        li2 = movie_dc[i]
        cos = dot(li1,li2)/(norm(li1)*norm(li2))
        simi.append(cos)
        mov.append(i)
    recomd_data = pd.DataFrame({'Movie':pd.Series(mov),'Similarity':pd.Series(simi)})
    return recomd_data.sort_values(by='Similarity',ascending=False).iloc[1:11,:]
```
